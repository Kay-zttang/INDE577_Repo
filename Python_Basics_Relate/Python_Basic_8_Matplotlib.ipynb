## Line and Scatter Plots
### Line Plots
- Create a line plot with plot() function

```python
import matplotlib.pyplot as plt
import numpy as np

def squared(x):
    return x**2

x = np.array([0.0, .5, 1.0, 1.5, 2.0])
y = squared(x)

print(f"{x = }")
print(f"{y = }")

plt.plot(x, y)
plt.title("My Title")
plt.show()
```

- Adjust the Figure Size

```python
plt.figure(figsize = (10,8))
plt.plot(x, y)
plt.show()
```

- Add Axis Labels / Titles / Legends

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, label = "f(x) = x^2")

plt.xlabel("time (s)")
plt.ylabel("distance (m)")

plt.legend(fontsize = 13)

plt.title("my First Plot", fontsize = 15)
plt.show()
```

### Scatter Plot
- Create scatter plot with scatter() function
- Adjustment same as Line Plot above

```python
plt.scatter(x, y, label = "points", color = "red")

plt.xlabel("time (s)")
plt.ylabel("distance (m)")

plt.legend(fontsize = 13)

plt.title("My First Scatter Plot")
plt.show()
```

### Overlaying Different Instances of Line and Scatter Plot

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, label = "f(x) = x^2", color = "green", linestyle = "--")
plt.scatter(x, y, label = "points", color ="red")

plt.xlabel("time (s)", fontsize = 12)
plt.ylabel("distance (m)")
plt.grid()
plt.legend()
plt.title("My First Mix Plot", fontsize = 15)
plt.show()
```

### Special Example of A Intricate Plot

```python
x = np.linspace(0, 2*np.pi, 100) 
# linspace() for evenly seperate a distance and return numbers
# pi() for showing the math symbol pi
special_x = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]

plt.figure(figsize = (10,8))
plt.plot(x, np.sin(x), label = "f(x) = sin(x)", color = "green", linestyle = "--")
plt.scatter(special_x, np.sin(special_x), color = "lightgreen")

plt.plot(x, np.cos(x), label = "f(x) = cos(x)", color = "red", linestyle ="-")
plt.scatter(special_x, np.cos(special_x), color = "magenta")

plt.grid()
plt.xticks(special_x, labels = ["0", "$\pi$/2", "$\pi$", "3$\pi$/2", "2$\pi$"],
          fontsize = 13)
plt.ylabel("distance (m)", fontsize = 13)
plt.legend()
plt.title("My Trig Plot", fontsize = 16)
plt.show()
```
